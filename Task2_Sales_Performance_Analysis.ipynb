```python
import pandas as pd
import matplotlib.pyplot as plt

from google.colab import files
uploaded = files.upload()

file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name, encoding='latin1')

# Cleaning
df.drop_duplicates(inplace=True)
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')
df['Ship_Date'] = pd.to_datetime(df['Ship_Date'], errors='coerce')

df['Year'] = df['Order_Date'].dt.year
df['Month'] = df['Order_Date'].dt.month
```

```python
print("Total Revenue:", df['Sales'].sum())
print("Average Order Value:", df['Sales'].mean())
if 'Profit' in df.columns:
    print("Total Profit:", df['Profit'].sum())
if 'Profit' in df.columns:
    total_profit = df['Profit'].sum()
    total_sales = df['Sales'].sum()
    print("Profit Margin (%):", (total_profit / total_sales) * 100)
```

```python
df.groupby('Region')['Sales'].sum().plot(kind='bar', title='Sales by Region')
plt.show()
```

```python
df.groupby('Category')['Sales'].sum().plot(kind='pie', autopct='%1.1f%%', title='Sales by Category')
plt.show()
```

```python
df.groupby('Year')['Sales'].sum().plot(marker='o', title='Yearly Sales Trend')
plt.show()
```

```python
df.groupby('Month')['Sales'].sum().plot(marker='o', title='Monthly Sales Trend')
plt.show()
```

```python
df.groupby('Product_Name')['Sales'].sum().sort_values(ascending=False).head(10).plot(kind='barh', title='Top 10 Products')
plt.show()
```

```python
df.groupby('Product_Name')['Sales'].sum().sort_values().head(10).plot(kind='barh', title='Worst 10 Products')
plt.show()
```
