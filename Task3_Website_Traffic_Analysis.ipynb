```python
# ==============================
# 📌 1. IMPORT LIBRARIES
# ==============================
import pandas as pd
import matplotlib.pyplot as plt

# ==============================
# 📌 2. UPLOAD FILE
# ==============================
from google.colab import files
uploaded = files.upload()

file_name = list(uploaded.keys())[0]

# ==============================
# 📌 3. LOAD DATA
# ==============================
df = pd.read_csv(file_name, encoding='latin1')

# ==============================
# 📌 4. DATA CLEANING
# ==============================
df.drop_duplicates(inplace=True)

# Check missing values
print("Missing Values:\n", df.isnull().sum())

# ==============================
# 📌 5. BASIC INFO
# ==============================
print("\nDataset Info:\n")
print(df.info())

# ==============================
# 📌 6. KPIs (METRICS)
# ==============================

# Total Sessions (rows count)
total_sessions = len(df)

# Total Users (if column exists)
if 'User_ID' in df.columns:
    total_users = df['User_ID'].nunique()
else:
    total_users = "Not Available"

# Bounce Rate (if exists)
if 'Bounce_Rate' in df.columns:
    bounce_rate = df['Bounce_Rate'].mean()
else:
    bounce_rate = "Not Available"

# Avg Session Duration (if exists)
if 'Session_Duration' in df.columns:
    avg_session = df['Session_Duration'].mean()
else:
    avg_session = "Not Available"

print("\nTotal Sessions:", total_sessions)
print("Total Users:", total_users)
print("Bounce Rate:", bounce_rate)
print("Avg Session Duration:", avg_session)

# ==============================
# 📌 7. TRAFFIC SOURCE ANALYSIS
# ==============================
if 'Traffic_Source' in df.columns:
    df['Traffic_Source'].value_counts().plot(kind='bar', title='Traffic by Source')
    plt.show()

# ==============================
# 📌 8. BOUNCE RATE DISTRIBUTION
# ==============================
if 'Bounce_Rate' in df.columns:
    df['Bounce_Rate'].plot(kind='hist', title='Bounce Rate Distribution')
    plt.show()

# ==============================
# 📌 9. SESSION DURATION
# ==============================
if 'Session_Duration' in df.columns:
    df['Session_Duration'].plot(kind='hist', title='Session Duration')
    plt.show()

# ==============================
# 📌 10. TOP LANDING PAGES
# ==============================
if 'Page' in df.columns:
    df['Page'].value_counts().head(10).plot(kind='barh', title='Top Landing Pages')
    plt.show()

# ==============================
# 📌 11. EXIT PAGES
# ==============================
# (Assuming Page column used for exits too)
if 'Page' in df.columns:
    df['Page'].value_counts().tail(10).plot(kind='barh', title='Exit Pages')
    plt.show()
```
